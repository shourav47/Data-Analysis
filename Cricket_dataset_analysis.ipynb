#Split a Column into Multiple Column

```python
df['Player'] = df['Player'].str.replace('(','')
df['Player'] = df['Player'].str.replace(')','')
df_player = df['Player'].str.split("\xa0",expand = True)
df_player.head(2)
```

###Concatation of two dataframe
<br>
----If we don't give axis = 1 then we get NAN values

```python
df_updated = pd.concat([df,df_player],axis = 1)
df_updated.head(5)
#df_updated['Country'].values
```

# Remove a column from a Dataset

```python
#method 1
#df.drop('column_name', axis=1, inplace=True)
#method 2
df_updated.drop(df_updated.columns[[0]], axis=1,inplace = True)
df_updated.head(5)
```

###Rename the column name of the dataset

```python
df_updated.rename(columns = {0:'Player', 1: 'Country'},inplace = True)
df_updated.head(3)
```

#Create a new column to check if the player have played for ICC or not

```python
def icc(x):
  if 'ICC/' in x:
    return 'Yes'
  else:
    return 'No'
df_updated['Played_for_ICC'] = df_updated['Country'].apply(icc)
df_updated.head(5)
```

###Remove unwanted ulaphabet from country column

```python
df_updated['Country'] = df_updated['Country'].str.replace('ICC/', '')
df_updated.head(5)
```

# Copy a dataset

####Rearranging the Column

```python
df_updated.columns
```

```python
df_colls = ['Player', 'Country','Span', 'Match', 'Innings', 'Not_Out', 'Runs', 'Highest_Score',
       'Average', 'Hundreds', 'Fifties', 'Duck','Played_for_ICC']
df_updated = df_updated[df_colls]
df_updated.head(5)
```

###Splitting the span column & find out years_played

```python
df_updated['Span'] = df_updated['Span'].str.replace('-','')
df_updated['Start_Year'] = df_updated['Span'].str[0:4]
df_updated['End_Year'] = df_updated['Span'].str[4:]
df_updated.head(5)
#df_updated['Span'].values
```

###Deleting Span column & make obj column to int

```python
#deleting object column
df_updated = df_updated.drop('Span', axis = 1)
df_updated.head(2)
```

```python
#make obj data into int
df_updated['Start_Year'] = df_updated['Start_Year'].astype('int')
df_updated['End_Year'] = df_updated['End_Year'].astype('int')
df_updated.info()
```

#Make a new column for years played 

```python
df_updated['Years_Played'] = df_updated['End_Year'] - df_updated['Start_Year']

#rearrenging the columns
df_colls = ['Player', 'Country', 'Years_Played','Match', 'Innings', 'Not_Out', 'Runs',
       'Highest_Score', 'Average', 'Hundreds', 'Fifties', 'Duck', 'Start_Year',
       'End_Year','Played_for_ICC']
df_updated = df_updated[df_colls]
df_updated.head(5)
```

#Removing star from the Highest score

```python
##mehod 1
# df_updated['Highest_Score'] = df_updated['Highest_Score'].astype('string')
# df_updated['Highest_Score'] = df_updated['Highest_Score'].str.replace('*','')
# df_updated['Highest_Score'] = df_updated['Highest_Score'].astype('int64')
```

```python
##method 2
def star_remover(x):
  x = str(x)
  if '*' in x:
    return x.replace('*','')
  else:
    return x
df_updated['Highest_Score'] = df_updated['Highest_Score'].apply(star_remover)
display(df_updated.head(5))
df_updated.info()
```

```python
df_updated['Highest_Score'] = df_updated['Highest_Score'].astype('int64')
df_updated.info()
```

#Removing the Start_Year & End_Year column

```python
df_updated = df_updated.drop(['Start_Year','End_Year'], axis =1)
df_updated.head(2)
df_updated.columns
```
