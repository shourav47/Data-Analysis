#Features or column description


*   **Player**: Name of the player & Country.
*   **Span**: Year period for a player to play cricket.
*  **Mat**: No of matches a player have played.
*  **Inns**: No of innings a player have played.
*   **NO**: No of matches for a player is Not Out.
*   **Runs**: Total runs a player have obtained during his entire career.
*   **Hs**: Highest score a player has obtained in a match.
*   **Ave**: Average run of a player in a match.
*   **100**: No of hundred a player has obtained in entire innigns.
*  **50**: No of fifties a player has obtained in entire innigns.
*   **0** : No of duck a player has obtained in entire innigns.
