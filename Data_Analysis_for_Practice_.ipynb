#Creating a Dataframe & Concatation of Dataframe

```python
#dataframe creation (method 1)
import pandas as pd
import numpy as np
student = pd.DataFrame(data={'Name':['A','B','C','D'],'Age':[11,22,34,45]})
#display(student)
```

```python
#dataframe creation (method 2)
teacher = pd.DataFrame()
teacher['Name']= ['c','d','e','f']
teacher['Age'] = [55,54,33,56]
#display(teacher)
```

```python
#dataframe concatation
school = pd.concat([student,teacher], axis=0)
# display(school.head(8))
school['a'] = np.arange(0,len(school),1)
school.set_index(school['a'],inplace= True)
display(school.head(8))
# s = pd.Series(np.arange(0,len(school),1))
# school.set_index([s])
# display(school)
```

```python
#dataframe concatation
school = pd.concat([student,teacher], axis=1)
display(school.head(8))
```

```python
#data frame cocatation with correct index
school = pd.concat([student, teacher], axis = 0)
school = school.set_index(np.arange(0,len(school),1))
display(school.head(10))
```

#Percentage of Missing values in each Columns

```python
# summarize the horse colic dataset
from pandas import read_csv
# load dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/horse-colic.csv'
dataframe = read_csv(url, header=None, na_values='?')
# summarize the first few rows
print(dataframe.head())
# summarize the number of rows with missing values for each column
for i in range(dataframe.shape[1]):
	# count number of rows with missing values
	n_miss = dataframe[[i]].isnull().sum()
	perc = n_miss / dataframe.shape[0] * 100
	print('> %d, Missing: %d (%.1f%%)' % (i, n_miss, perc))
```
