#Creating a Dataframe & Concatation of Dataframe

```python
#dataframe creation (method 1)
import pandas as pd
import numpy as np
student = pd.DataFrame(data={'Name':['A','B','C','D'],'Age':[11,22,34,45]})
#display(student)
```

```python
#dataframe creation (method 2)
teacher = pd.DataFrame()
teacher['Name']= ['c','d','e','f']
teacher['Age'] = [55,54,33,56]
#display(teacher)
```

```python
#dataframe concatation
school = pd.concat([student,teacher], axis=0)
# display(school.head(8))
school['a'] = np.arange(0,len(school),1)
school.set_index(school['a'],inplace= True)
display(school.head(8))
# s = pd.Series(np.arange(0,len(school),1))
# school.set_index([s])
# display(school)
```

```python
#dataframe concatation
school = pd.concat([student,teacher], axis=1)
display(school.head(8))
```

```python
#data frame cocatation with correct index
school = pd.concat([student, teacher], axis = 0)
school = school.set_index(np.arange(0,len(school),1))
display(school.head(10))
```

#Percentage of Missing values in each Columns

```python
# summarize the horse colic dataset
from pandas import read_csv
# load dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/horse-colic.csv'
dataframe = read_csv(url, header=None, na_values='?')
# summarize the first few rows
print(dataframe.head())
# summarize the number of rows with missing values for each column
for i in range(dataframe.shape[1]):
	# count number of rows with missing values
	n_miss = dataframe[[i]].isnull().sum()
	perc = n_miss / dataframe.shape[0] * 100
	print('> %d, Missing: %d (%.1f%%)' % (i, n_miss, perc))
```

#Advanced iloc & loc practice in python

```python
# crete a sample dataframe
data = pd.DataFrame({
    'age' :     [ 10, 22, 13, 21, 12, 11, 17],
    'section' : [ 'A', 'B', 'C', 'B', 'B', 'A', 'A'],
    'city' :    [ 'Gurgaon', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai'],
    'gender' :  [ 'M', 'F', 'F', 'M', 'M', 'M', 'F'],
    'favourite_color' : [ 'red', np.NAN, 'yellow', np.NAN, 'black', 'green', 'red']})
display(data)
```

###Question on mind
1.  Find the values of the rows where age is greater than 15

```python
display(data.loc[data.age>=15])
```

2.  Finding all the rows where the age is greater than or equal to 12 and the gender is also male:

```python
display(data.loc[(data.age >=12) & (data.gender == 'M')])
```

3.  Find the last 3 rows with last 3 column using loc & iloc methohd

```python
#loc depends on the index position. It gives only 4,5,6 rows.
display(data.loc[4:6])

#iloc maintains indexing. & upper limit excluded.
display(data.iloc[4:6,2:5])
```

4.  Select the column Gender & City where age is greater than 13.

```python
display(data.loc[(data.age>=13),['city','gender']])
```

#Update the gender 'M' to 'F' when the age is greater than 15

```python
data.loc[(data.age>=13), ['gender']] = 'F'
display(data)
```

```python
# update multiple columns with condition
data.loc[(data.age >= 20), ['section', 'city']] = ['S','Pune']
data
```

```python
# select rows with indexes
display(data.iloc[0:2,:])
```

```python
#showing difference between df1 & df2
df1 = data.iloc[[0,2],[1,3]]
display(df1)
df2 = data.iloc[0:2,1:3]
display(df2)
```
